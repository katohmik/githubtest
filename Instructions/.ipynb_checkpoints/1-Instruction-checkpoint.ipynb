## (5) Branch
	
   - ※ 소스 코드 관리(subversion)
   - 브랜치는 커밋을 가리키는 포인터(pointer)
   - Git 브랜치는 특정 시점에서의 프로젝트 변경이력을 기록하는 스냅숏
   - branches (기능 확장, 수정을 위한 임시 작업실)
   - tags
   - trunk

- <tag<tag>> 위치를 언제나 파악해야한다

    * : 현재의 head를 나타낸다.
    *master : 헤드가 마스터에 있다.
 
- git branch : 브랜치 정보 보기
- git branch <브랜치명> : 브랜치 생성
- git checkout <브랜치명> : 브랜치 전환

- git log --online : 한줄씩 간단하게 로그를 확인가능
- git log --decorate --graph --all : 그림으로 선을 보여줌
- git branch --delete <브랜치명> 또는 git branch -D <브랜치명>

#### 1) git pull <다운로드할 서버 주소> <데이터를 받을 장소>
> - pull : fetch + merge
		------------------------------------------------------------
> - git pull origin master
    -받아서 바로 합치자
    ※ 다른 사람이 올린 새 파일이 있을 경우, 우선 그 파일을 다운받아야 자신의 것을 올릴 수 있다

#### 2) git fetch <다운로드할 서버 주소> <데이터를 받을 주소>
> - git merge <합칠 데이터 주소>
> 따로 합치는 작업을 해줘야하는 fetch, 우선 받아만 놓고 확인한 뒤에 합치자.
>		-------------------------------------------------------------------------------
>> - git fetch origin master
>> - git checkout FETCH_HEAD
>> - FETCH_HEAD로 들어오면 아까는 보이지 않던 파일이 보이고, 여기서 문제가 있는지 확인이 가능
>> - 다시 master로 돌아가야함
>> - git checkout master
>> - master로 돌아오니까 다시 안보임
>>	- git merge FETCH_HEAD

#### 3) git clone <서버주소>
   - 서버에 있는 레포지토리를 그래도 다운받을 수 있음
   - github 레포지토리에서 code 누르고 서버주소 복사해서 넣어줌(보통 관리자가 알려줌)

# 2. Git 협업

## (1) Git Branch(협업) 사용법

- 누군가가 만든 레포지토리에 내 브런치가 없는 경우에 그곳에 파일을 올리고싶을 경우

1. git bash 접속
2. git remote 확인 및 접속
2. git clone 협업할 레포지토리 주소
3. cd 클론된 폴더 이름(이곳으로 반드시 이동)
4. git branch(branch상태 확인, master일경우 개인 브런치로 이동필요)
5. git checkout 개인브런치이름
6. git branch(branch상태 확인, 개인브런치인지)
7. 클론된 폴더 안에 내가 작업한 파일을 넣어줌
8. git add
9. git status(파일이 정확하게 올라갔는지 확인)
10. git commit -m "commit 메세지 입력"
11. git status(확인용, 생략가능)
12. git push
13. 4~5에서 만들고 이동한 branch명으로 레포지토리에 올라갔는지 확인

## (2) Git Branch(협업) 사용법

- 누군가가 만든 레포지토리에 내 브런치가 있는 경우에 그곳에 파일을 올리고싶을 경우

1. git bash 접속
2. git remote 확인 및 접속
2. git clone 협업할 레포지토리 주소
3. cd 클론된 폴더 이름(이곳으로 반드시 이동)

6. git branch(branch상태 확인, 레포지토리에 있는 업로드하려는 브랜치명과 동일한지)
7. 클론된 폴더 안에 내가 작업한 파일을 넣어줌
8. git add
9. git status(파일이 정확하게 올라갔는지 확인)
10. git commit -m "commit 메세지 입력"
11. git push
12. git status(확인용, 생략가능)
13. 같은 local PC이면서 초기화하지 않은 경우 + branch가 아직 있을 경우 = 4부터 시작해도 괜찮다
14. 여러개의 리모트를 옮겨다니며 사용할경우에는 안전하게 2번부터 확인하는 습관이 필요하다

## (4) Git 로컬 branch, remote branch 삭제하기

### 1) branch를 삭제하는 이유

프로젝트에서 코드의 변경 사항을 저장하기 위해 브랜치를 만들지만 해당 브랜치의 변경 사항이나 새 기능을 프로젝트의 본 버전에 반영 완료한 후에는 해당 브랜치를 유지할 필요가 없으므로, 프로젝트 저장소가 복잡해지지 않도록 작업이 완료된 브랜치는 삭제하는 것이 일반적.

### 2) Local Branch 삭제하기

> git checkout <브랜치 이름>

- local branch에서 작업중이라면 삭제가 안되므로, 다른 branch로 이동필수(보통 main)

> git branch -d <로컬 브랜치 이름>

- git branch는 로컬 브랜치를 삭제하는 명령어입니다.
- -d는 --delete의 약자, 선택적 플래그
- <로컬 브랜치 이름>은 삭제하고 싶은 로컬 브랜치를 가리킵니다.
- 브랜치에 병합되지 않은 변경사항, 푸시되지 않은 커밋이 있는 경우 -d 플래그를 사용불가

> git branch -D <로컬 브랜치 이름>

- 대문자"D"의 -D 플래그는 --delete --force(강제 삭제)의 줄임말이며, 병합 여부와 관계없이 로컬 브랜치를 강제로 삭제

### 3) Remote Branch 삭제하기

- 원격 브랜치와 로컬 브랜치는 독립적인 별개의 개체
- 본인 컴퓨터의 로컬 브랜치와 달리, 원격 브랜치는 원격 서버에 저장되어 있으며 원격 서버를 통해 접근가능

> git push <원격 저장소 이름> -d <원격 브랜치 이름>

- 로컬 브랜치를 삭제하는 git branch 명령어를 사용하는 대신 git push 명령어를 이용해 원격 브랜치를 삭제
- git push 명령어 다음에 원격 저장소의 이름을 전달해야 합니다(대부분 origin)
- <원격 브랜치 이름>은 삭제하고 싶은 원격 브랜치

> git branch -a

- 모든 브랜치 확인(local + remote)
- -a = --all(의 줄임말)

> git branch -r

- 모든 원격 브랜치만 확인
