```python
#train[train["Survived"]==1]
survived = train[train["Survived"]==1]["Sex"].value_counts()
dead = train[train["Survived"]==0]["Sex"].value_counts()

df = pd.DataFrame([survived,dead])
df.index = ["Survived", "Dead"]
df.plot(kind="bar", stacked=True, figsize=(5, 3))
```

```python
def bar_chart(feature):
    survived = train[train["Survived"]==1][feature].value_counts()
    dead = train[train["Survived"]==0][feature].value_counts()

    df = pd.DataFrame([survived,dead])
    df.index = ["Survived", "Dead"]
    df.plot(kind="bar", stacked=True, figsize=(5, 3))
```

### (3) Feature Engineering (특성 공학, 전처리 - ETL(Extrack Transform Load))

```python
train["Title"] = train["Name"].str.extract(" ([a-zA-Z]+[.])")
train.head(10)
```

```python
train["Name"].str.extract(" ([a-zA-Z]+[.])")[0].unique()
```

```python
train["Title"].value_counts()
```

```python
title_map = {'Mr.':0, 'Mrs.':1, 'Miss.':2, 'Master.':3, 'Don.':3, 'Rev.':3, 'Dr.':3, 'Mme.':3,
       'Ms.':3, 'Major.':3, 'Lady.':3, 'Sir.':3, 'Mlle.':3, 'Col.':3, 'Capt.':3,
       'Countess.':3, 'Jonkheer.':3}

train["Title1"] = train["Title"].map(title_map)
```

```python
train.head()
```

```python
bar_chart("Title1")
```

#### 2) Age 결측치 처리

```python
# 결측치를 대체하자

train.groupby("Title1")["Age"].median() # 중앙값을 찾다
```

```python
train["Age"].fillna(train.groupby("Title1")["Age"].apply("median")) # apply는 그룹으로 묶어서 대응하므로 NaN이 그대로임
```

```python
train["Age"].fillna(train.groupby("Title1")["Age"].transform("median")) # transform은 하나하나 대응하므로 NaN이 사라짐.
```

```python
train["Age"].fillna(train.groupby("Title1")["Age"].transform("median"),inplace=True)
```

## 11. WordCloud의 한글 처리

```python
import konlpy
from konlpy.tag import Okt
from konlpy.corpus import kolaw
import nltk
from wordcloud import WordCloud
```

```python
okt = Okt()
okt.nouns("아버지가방에들어가신다.")
```

```python
##### 형태소 분석

ko_text=kolaw.open("constitution.txt").read()
ko_text
```

```python
token_ko = okt.nouns(ko_text)
token_ko

print(len(token_ko))
```

```python
##### 불용어 처리

stop_words = ["제", "월", "일", "조", "수", "때", "그", "아", "이", "바", "및", "안", "정", "위", "관", "이하", "아래"]
```

```python
token_ko = [word for word in token_ko if word not in stop_words]
```

```python
print(len(token_ko))
```

```python
##### 빈도수 처리(워드 카운팅)

ko = nltk.Text(token_ko, name = "대한민국 헌법")
dir(ko)

data = ko.vocab().most_common(150)
```

```python
wc = WordCloud(max_font_size=40, background_color="black",
                max_words=200,
              font_path="C:\\Windows\\Fonts\\malgun.ttf").generate_from_frequencies(dict(data))

plt.figure(figsize=(12,12))
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
```
