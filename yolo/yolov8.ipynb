## yolov8 YAML 설정

```python
!pip install PyYAML
```

```python
import os
import yaml

data = {
    "train": "/content/train/images/",
    "val": "/content/valid/images/",
    "test": "/content/test/images/",
    "nc": 7,
    "names": ['fish', 'jellyfish', 'penguin', 'puffin', 'shark', 'starfish', 'stingray']
}

# directory = "/content/aquarium_data"
# if not os.path.exists(directory):
#     os.makedirs(directory)

with open("/content/aquarium_data.yaml", "w") as f:
    yaml.dump(data, f)
```

```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")
```

```python
model.train(data="/content/aquarium_data.yaml", epochs=100, patience=30, batch=32, imgsz=416)
```

```python
results = model.predict(source="/content/test/images/", save=True)
```
